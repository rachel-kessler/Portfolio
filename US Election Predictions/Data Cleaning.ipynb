## Data Cleaning

After the intial exploration, some steps to clean the data will take place:
### Removing Duplicates
A heatmap will be used to quickly identify columns which may be exact duplicates

```python
plt.figure(figsize = (16,16))
sns.heatmap(df_data.corr(), cmap="coolwarm")
```

It appears that there are duplicated ethnicity columns. This will be confirmed by looking at exact correlation values.

```python
df_data.corr()
```

It is confirmed that the ethnicity columns are duplicated and neither appear to have NaN values. The second set will be deleted from the dataset

```python
columns = df_data.loc[:,'White':'White_Asian']
df_data.drop(columns, inplace=True, axis=1)
```

The ST column is also a duplicate and will be removed:

```python
df_data.drop('ST', inplace=True, axis=1)
```

```python
df_data.info()
```

### Cleaning Null Values
Null values were then examined in the **votes** dataframe:

```python
df_votes[df_votes.isna().any(axis=1)]
```

Some rows have no data input; these rows will be dropped from the votes table.

```python
df_votes.dropna(thresh=6, inplace=True)
df_votes.info()
```

32 Rows were dropped from the table which contained no information.

Null values were then examined in the **data** dataframe:

```python
df_data.info()
```

```python
df_data.head()
```

Infant Mortality and Homicide Rate contain less than 50% of the values so these columns will be removed.

The final 14 columns appear to contain the most Null Values

```python
df_data.drop('Homicide.rate', axis=1, inplace=True)
```

```python
df_data.drop('Infant.mortality', axis=1, inplace=True)
```

```python
df_data.loc[:,'Low.birthweight':].describe()
```

It appears that the mean and median values for these columns are generally similar, except in the case of sexually transmitted infections, HIV prevalence, and violent crime rates (suggesting these may be skewed). The median values will be used to fill in null values in each of these columns to avoid any effects of skewed data. 

```python
df_data.loc[:,'Low.birthweight':] = df_data.loc[:,'Low.birthweight':].fillna(df_data.median())
df_data.info()
```

```python
df_data.loc[:,['Child.Poverty.living.in.families.below.the.poverty.line','Preschool.Enrollment.Ratio.enrolled.ages.3.and.4']].describe()
```

Preschool enrollment and child poverty also contain a handful of null values. These will also be filled in using the median 

```python
df_data.loc[:,['Child.Poverty.living.in.families.below.the.poverty.line','Preschool.Enrollment.Ratio.enrolled.ages.3.and.4']]\
= df_data.loc[:,['Child.Poverty.living.in.families.below.the.poverty.line','Preschool.Enrollment.Ratio.enrolled.ages.3.and.4']]\
.fillna(df_data.median())
df_data.info()
```

The final cells which are missing values include 32 rows of precincts and votes:

```python
df_data[df_data.isna().any(axis=1)]
```

These Fips codes were deleted from the df_votes dataframe because they contained null values. They will be deleted from this dataset as well.

```python
df_data.dropna(inplace=True)
```

```python
df_data.info()
```

The data has been cleaned and each dataset contains information about 3109 counties in the United States.

### Relative Vote Count
The final step in preparing the data for further analysis was adding relative vote counts for each county in the votes dataframe

```python
df_votes.head()
```

```python
df_votes['Dem_2008_Share'] = df_votes['Democrats 08 (Votes)']/(df_votes['Democrats 08 (Votes)']+df_votes['Republicans 08 (Votes)'])*100
df_votes['Rep_2008_Share'] = df_votes['Republicans 08 (Votes)']/(df_votes['Democrats 08 (Votes)']+df_votes['Republicans 08 (Votes)'])*100

df_votes['Dem_2012_Share'] = df_votes['Democrats 08 (Votes)']/(df_votes['Democrats 12 (Votes)']+df_votes['Republicans 12 (Votes)'])*100
df_votes['Rep_2012_Share'] = df_votes['Republicans 08 (Votes)']/(df_votes['Democrats 12 (Votes)']+df_votes['Republicans 12 (Votes)'])*100

df_votes['Dem_2016_Share'] = df_votes['votes16_clintonh']/(df_votes['votes16_clintonh']+df_votes['votes16_trumpd'])*100
df_votes['Rep_2016_Share'] = df_votes['votes16_trumpd']/(df_votes['votes16_clintonh']+df_votes['votes16_trumpd'])*100
```

```python
df_votes.head()
```

The relative vote shares for each party in each election have been added and the data is ready for further analysis.
